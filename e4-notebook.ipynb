<h2>1.</h2>

```python
# a)

d1 = P.Series([2, 3, 5, 4, 4, 3])
d2 = P.Series([4, 2, 3, 5, 2, 3])
d3 = P.Series([3, 1, 4, 4, 3, 5])
```

```python
# normality test with shapiro-wilk test

p_d1 = S.shapiro(d1).pvalue
print(p_d1)

p_d2 = S.shapiro(d2).pvalue
print(p_d2)

p_d3 = S.shapiro(d3).pvalue
print(p_d3)
```

```python
# all datasets are similar and normally distributed -> ANOVA

# first create dataframe

df1 = P.concat([d1, d2, d3], axis=1)
df1.columns = ['d1', 'd2', 'd3']
df1
```

```python
# modify dataframe into readable form for statsmodel ANOVA

modified_df1 = P.melt(df1,value_vars=['d1','d2','d3'])
modified_df1.columns = ['dataset','value']
modified_df1
```

```python
# define model for ANOVA

model1 = statsmodels.formula.api.ols('value~C(dataset)',data=modified_df1).fit()
```

```python
# finally we can perform ANOVA test

anova1 = statsmodels.api.stats.anova_lm(model1,typ=1)
anova1
```

```python
# b)

n1 = P.Series(['Fi', 'Sw', 'Fi', 'No', 'Sw', 'Fi'])
n2 = P.Series(['No', 'Sw', 'No', 'Fi', 'Fi', 'Fi'])
n3 = P.Series(['Sw', 'Fi', 'No', 'Sw', 'Sw', 'No'])
```

```python
# modify catogorical data into readable form for Pearson's chi-squared test

# Create a DataFrame
nf = P.DataFrame({'n1': n1, 'n2': n2, 'n3': n3})

# Count occurrences of each category by dataset
result_nf = nf.apply(lambda col: col.value_counts()).astype(int)

# Rename the index to 'Fi', 'Sw', 'No'
result_nf.index = ['Fi', 'No', 'Sw']

# Rename the columns to 'n1', 'n2', 'n3'
result_nf.columns = ['n1', 'n2', 'n3']

# Print the result
result_nf
```

```python
# Pearson's chi-squared test

S.chi2_contingency(result_nf)
```

```python
# load datafile

df3 = P.read_csv('./simulated_data_2_5.csv',
           # field separator character
           sep=",",
           # missing value characters
           na_values=["?"],
           # no row names
           index_col=0,
           # no column names
           header=0,
           # quote character
           quotechar="\"")
df3
```

```python
# Shapiro-Wilk tests

p_F = S.shapiro(df3['F']).pvalue
print(p_F)

p_G = S.shapiro(df3['G']).pvalue
print(p_G)

p_H = S.shapiro(df3['H']).pvalue
print(p_H)

p_I = S.shapiro(df3['I']).pvalue
print(p_I)
```

```python
# a)

# we can assume columns F, G and H are all normally distributed -> ANOVA

# all datasets are similar and normally distributed -> ANOVA

# first create dataframe

df3a = df3[['F','G','H']]
df3a
```

```python
# modify dataframe into readable form for statsmodel ANOVA

modified_df3a = P.melt(df3a,value_vars=['F','G','H'])
modified_df3a.columns = ['dataset','value']
modified_df3a
```

```python
# define model for ANOVA

model3 = statsmodels.formula.api.ols('value~C(dataset)',data=modified_df3a).fit()
```

```python
# finally we can perform ANOVA test

anova3 = statsmodels.api.stats.anova_lm(model3,typ=1)
anova3
```

```python
# b)

# I is not normally distributed -> Kruskal-Wallis test

S.kruskal(df3['F'], df3['G'], df3['I'])
```

<h2>5.</h2>

```python
# load datafiles

df5 = P.read_csv('./horse-colic.data.csv',
           # field separator character
           sep="\s+",
           # missing value characters
           na_values=["?"],
           # no row names
           index_col=False,
           # no column names
           header=None)
```

```python
df5 = df5.rename(columns={0: 'surgery', 1: 'age', 3: 'rectal temperature', 4: 'pulse'})
df5
```

```python
# groups

grouped = df5.groupby('surgery')

sur_true = grouped.get_group(1)

sur_false = grouped.get_group(2)
```

```python
# normality tests

print(S.shapiro(sur_true['rectal temperature'].dropna()).pvalue)
print(S.shapiro(sur_true['age'].dropna()).pvalue)
print(S.shapiro(sur_true['pulse'].dropna()).pvalue)
print(S.shapiro(sur_false['rectal temperature'].dropna()).pvalue)
print(S.shapiro(sur_false['age'].dropna()).pvalue)
print(S.shapiro(sur_false['pulse'].dropna()).pvalue)

print('\nNone of the variables are normally distributed')
```

```python
# all sets of samples are unpaired and not normally distributed

# can use Mann-Whitney U test for for all testing

# rectal temperature

S.mannwhitneyu(sur_true['rectal temperature'].dropna(), sur_false['rectal temperature'].dropna(),\
               use_continuity=True, alternative='two-sided')
```

```python
# age

S.mannwhitneyu(sur_true['age'].dropna(), sur_false['age'].dropna(),\
               use_continuity=True, alternative='two-sided')
```

```python
# pulse

S.mannwhitneyu(sur_true['pulse'].dropna(), sur_false['pulse'].dropna(),\
               use_continuity=True, alternative='two-sided')
```

We can conclude that both 'rectal temperature' and 'age' are not statistically significantly 
different between horses treated with surgery and horses treated without surgery.

However, 'pulse' is statistically significantly different between horse treated with
surgery and horses treated without surgery.
